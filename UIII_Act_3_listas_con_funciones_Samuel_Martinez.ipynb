#### LISTAS CIENTIFICO

```python
# Nuestra base de datos (Lista)
lista_cientificos = []

# --- DEFINICIÓN DE LAS 5 FUNCIONES ---

def agregarcientifico(quien, nombre):
    lista_cientificos.append(nombre)
    print(f"\n>>> [{quien}]: He agregado a '{nombre}' a la base de datos.")

def listarcientifico(quien):
    print(f"\n>>> [{quien}]: MOSTRANDO LISTA ACTUAL:")
    if not lista_cientificos:
        print("La lista está vacía actualmente.")
    else:
        for i, nombre in enumerate(lista_cientificos, 1):
            print(f"{i}. {nombre}")

def buscarcientifico(quien, nombre):
    if nombre in lista_cientificos:
        print(f"\n>>> [{quien}]: ¡Localizado! '{nombre}' sí está en la lista.")
    else:
        print(f"\n>>> [{quien}]: No se encontró a '{nombre}'.")

def modificarcientifico(quien, antiguo, nuevo):
    if antiguo in lista_cientificos:
        indice = lista_cientificos.index(antiguo)
        lista_cientificos[indice] = nuevo
        print(f"\n>>> [{quien}]: Éxito. Se cambió '{antiguo}' por '{nuevo}'.")
    else:
        print(f"\n>>> [{quien}]: Error. '{antiguo}' no existe para ser modificado.")

def borrarcientifico(quien, nombre):
    if nombre in lista_cientificos:
        lista_cientificos.remove(nombre)
        print(f"\n>>> [{quien}]: Registro eliminado: '{nombre}'.")
    else:
        print(f"\n>>> [{quien}]: Error. No se pudo borrar porque '{nombre}' no existe.")

# --- MENÚ INTERACTIVO ---

def ejecutar_sistema():
    print("--- REGISTRO DE CIENCIA ---")
    # Identificamos qué científico está operando el sistema
    usuario = input("¿Quién opera el sistema? (ej. Cientifico1): ")

    while True:
        print("\n" + "="*35)
        print(f" USUARIO ACTIVO: {usuario}")
        print("="*35)
        print("1. Agregar científico")
        print("2. Listar científicos")
        print("3. Buscar científico")
        print("4. Modificar científico")
        print("5. Borrar científico")
        print("6. Salir")

        opcion = input("\nElige una opción (1-6): ")

        if opcion == "1":
            nombre = input("Nombre del científico a registrar: ")
            agregarcientifico(usuario, nombre)

        elif opcion == "2":
            listarcientifico(usuario)

        elif opcion == "3":
            nombre = input("Nombre que deseas buscar: ")
            buscarcientifico(usuario, nombre)

        elif opcion == "4":
            antiguo = input("Nombre que quieres cambiar: ")
            nuevo = input("Nuevo nombre corregido: ")
            modificarcientifico(usuario, antiguo, nuevo)

        elif opcion == "5":
            nombre = input("Nombre que vas a eliminar: ")
            borrarcientifico(usuario, nombre)

        elif opcion == "6":
            print(f"Finalizando sesión. ¡Buen trabajo, {usuario}!")
            break

        else:
            print("Opción inválida. Por favor, marca un número del 1 al 6.")

# Iniciar el programa
if __name__ == "__main__":
    ejecutar_sistema()
```
