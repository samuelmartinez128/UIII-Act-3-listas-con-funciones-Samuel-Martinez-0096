#### LISTAS CIENTIFICO

```python
# Nuestra base de datos (Lista)
lista_cientificos = []

# --- DEFINICIÓN DE LAS FUNCIONES ---

def agregarcientifico(quien, nombre):
    lista_cientificos.append(nombre)
    print(f"\n>>> [{quien}]: He agregado a '{nombre}' a la lista.")

def leercientifico(quien):
    # Función original solicitada
    print(f"\n>>> [{quien}]: LEYENDO DATOS... La lista es: {lista_cientificos}")

def listarcientifico(quien):
    # Función agregada (nueva)
    print(f"\n>>> [{quien}]: LISTADO DETALLADO:")
    if not lista_cientificos:
        print("La lista está vacía.")
    else:
        for i, nombre in enumerate(lista_cientificos, 1):
            print(f"{i}. {nombre}")

def buscarcientifico(quien, nombre):
    if nombre in lista_cientificos:
        print(f"\n>>> [{quien}]: ¡Encontrado! '{nombre}' está registrado.")
    else:
        print(f"\n>>> [{quien}]: '{nombre}' no se encuentra en la base de datos.")

def modificarcientifico(quien, antiguo, nuevo):
    if antiguo in lista_cientificos:
        indice = lista_cientificos.index(antiguo)
        lista_cientificos[indice] = nuevo
        print(f"\n>>> [{quien}]: Modificación exitosa: '{antiguo}' ahora es '{nuevo}'.")
    else:
        print(f"\n>>> [{quien}]: Error al modificar. '{antiguo}' no existe.")

def borrarcientifico(quien, nombre):
    if nombre in lista_cientificos:
        lista_cientificos.remove(nombre)
        print(f"\n>>> [{quien}]: Se ha eliminado a '{nombre}' de la lista.")
    else:
        print(f"\n>>> [{quien}]: Error al borrar. '{nombre}' no existe.")

# --- MENÚ INTERACTIVO ---

def menu_principal():
    print("--- SISTEMA DE GESTIÓN CIENTÍFICA ---")
    usuario = input("Ingrese el nombre del científico que opera (ej. Cientifico1): ")

    while True:
        print("\n" + "—"*40)
        print(f" OPERADOR ACTUAL: {usuario}")
        print("—"*40)
        print("1. Agregar científico")
        print("2. Leer científicos (Formato rápido)")
        print("3. Listar científicos (Formato numerado)")
        print("4. Buscar científico")
        print("5. Modificar científico")
        print("6. Borrar científico")
        print("7. Salir")

        opcion = input("\nSeleccione una acción (1-7): ")

        if opcion == "1":
            nombre = input("Nombre a agregar: ")
            agregarcientifico(usuario, nombre)

        elif opcion == "2":
            leercientifico(usuario)

        elif opcion == "3":
            listarcientifico(usuario)

        elif opcion == "4":
            nombre = input("Nombre a buscar: ")
            buscarcientifico(usuario, nombre)

        elif opcion == "5":
            antiguo = input("Nombre actual: ")
            nuevo = input("Nombre nuevo: ")
            modificarcientifico(usuario, antiguo, nuevo)

        elif opcion == "6":
            nombre = input("Nombre a eliminar: ")
            borrarcientifico(usuario, nombre)

        elif opcion == "7":
            print(f"Cerrando sistema. ¡Hasta pronto, {usuario}!")
            break

        else:
            print("Opción no válida, intente de nuevo.")

# Ejecución del programa
if __name__ == "__main__":
    menu_principal()
```
