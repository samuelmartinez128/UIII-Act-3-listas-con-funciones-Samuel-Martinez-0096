#### LISTAS CIENTIFICO

```python
# Nuestra base de datos (Lista)
lista_cientificos = []

# --- DEFINICIÓN DE LAS 5 FUNCIONES ---

def agregarcientifico(quien, nombre):
    lista_cientificos.append(nombre)
    print(f"\n>>> [{quien}]: Se ha registrado a '{nombre}' con éxito.")

def leercientifico(quien):
    print(f"\n>>> [{quien}]: CONSULTA DE LISTA:")
    if not lista_cientificos:
        print("La lista está vacía.")
    else:
        for i, nombre in enumerate(lista_cientificos, 1):
            print(f"{i}. {nombre}")

def buscarcientifico(quien, nombre):
    if nombre in lista_cientificos:
        print(f"\n>>> [{quien}]: ¡ÉXITO! '{nombre}' sí está en el sistema.")
    else:
        print(f"\n>>> [{quien}]: ERROR: '{nombre}' no fue encontrado.")

def modificarcientifico(quien, antiguo, nuevo):
    if antiguo in lista_cientificos:
        indice = lista_cientificos.index(antiguo)
        lista_cientificos[indice] = nuevo
        print(f"\n>>> [{quien}]: CAMBIO REALIZADO. '{antiguo}' ahora es '{nuevo}'.")
    else:
        print(f"\n>>> [{quien}]: ERROR: No se puede modificar porque '{antiguo}' no existe.")

def borrarcientifico(quien, nombre):
    if nombre in lista_cientificos:
        lista_cientificos.remove(nombre)
        print(f"\n>>> [{quien}]: ELIMINADO. '{nombre}' ha sido borrado.")
    else:
        print(f"\n>>> [{quien}]: ERROR: '{nombre}' no existe en la lista.")

# --- MENÚ INTERACTIVO ---

def menu():
    # Identificamos quién está usando el programa
    print("--- ACCESO AL SISTEMA ---")
    usuario = input("Ingrese su nombre (ej. Cientifico1): ")

    while True:
        print("\n" + "="*30)
        print(f" SESIÓN: {usuario}")
        print("="*30)
        print("1. Agregar científico")
        print("2. Ver lista de científicos")
        print("3. Buscar científico")
        print("4. Modificar científico")
        print("5. Borrar científico")
        print("6. Salir")

        opcion = input("\nSeleccione una opción (1-6): ")

        if opcion == "1":
            nombre = input("Nombre del nuevo científico: ")
            agregarcientifico(usuario, nombre)

        elif opcion == "2":
            leercientifico(usuario)

        elif opcion == "3":
            nombre = input("Nombre a buscar: ")
            buscarcientifico(usuario, nombre)

        elif opcion == "4":
            antiguo = input("Nombre actual que desea cambiar: ")
            nuevo = input("Nuevo nombre: ")
            modificarcientifico(usuario, antiguo, nuevo)

        elif opcion == "5":
            nombre = input("Nombre a eliminar: ")
            borrarcientifico(usuario, nombre)

        elif opcion == "6":
            print(f"Cerrando sesión de {usuario}. ¡Adiós!")
            break

        else:
            print("Opción no válida, intente de nuevo.")

# Ejecutar el programa
menu()
```
