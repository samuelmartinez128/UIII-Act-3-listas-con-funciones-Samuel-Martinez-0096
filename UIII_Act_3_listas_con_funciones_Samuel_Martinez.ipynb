#### CLASE CIENTIFICO

```python
class Cientifico:
    def __init__(self, nombre, clave, especialidad):
        # 3 Atributos
        self.nombre = nombre
        self.clave = clave
        self.especialidad = especialidad

    # --- Las 5 funciones (acciones) ---
    def agregarcientifico(self):
        print(f"\n[SISTEMA]: Científico {self.nombre} registrado con éxito.")

    def leercientifico(self):
        print(f"\n--- Datos del Científico ---")
        print(f"Nombre: {self.nombre}")
        print(f"Clave: {self.clave}")
        print(f"Especialidad: {self.especialidad}")

    def buscarcientifico(self):
        print(f"\n[BUSCADOR]: Localizando clave {self.clave}...")
        print(f"Resultado: El científico {self.nombre} se encuentra en la base de datos.")

    def modificarcientifico(self):
        print(f"\nModificando datos de: {self.nombre}")
        self.nombre = input("Ingrese el nuevo nombre: ")
        self.especialidad = input("Ingrese la nueva especialidad: ")
        print("[SISTEMA]: Registro actualizado.")

    def borrarcientifico(self):
        print(f"\n[ELIMINANDO]: Borrando registro de {self.nombre}...")
        self.nombre = "Vacío"
        self.clave = "0"
        self.especialidad = "N/A"
        print("[SISTEMA]: Los datos han sido borrados.")

# --- Instanciación de los 3 objetos (Científicos) ---
# Se inicializan con datos genéricos para que el usuario los llene
cientifico1 = Cientifico("Pendiente", "001", "N/A")
cientifico2 = Cientifico("Pendiente", "002", "N/A")
cientifico3 = Cientifico("Pendiente", "003", "N/A")

# --- Menú interactivo con While, If, Elif, Break ---
while True:
    print("\n====================================")
    print("   GESTIÓN DE CIENTÍFICOS - MENÚ    ")
    print("====================================")
    print("1. Agregar datos (Científico 1)")
    print("2. Leer datos (Científico 1)")
    print("3. Buscar científico (Científico 2)")
    print("4. Modificar datos (Científico 3)")
    print("5. Borrar datos (Científico 1)")
    print("6. SALIR")

    opcion = input("\nSeleccione una opción: ")

    if opcion == "1":
        print("\n--- Registro Científico 1 ---")
        nom = input("Nombre: ")
        cla = input("Clave: ")
        esp = input("Especialidad: ")
        cientifico1 = Cientifico(nom, cla, esp)
        cientifico1.agregarcientifico()

    elif opcion == "2":
        cientifico1.leercientifico()

    elif opcion == "3":
        # Simulamos búsqueda con el objeto 2
        cientifico2.buscarcientifico()

    elif opcion == "4":
        # Modificamos el objeto 3
        cientifico3.modificarcientifico()

    elif opcion == "5":
        cientifico1.borrarcientifico()

    elif opcion == "6":
        print("\nCerrando programa...")
        break # Detiene el ciclo While

    else:
        print("\n[ERROR]: Opción no válida, intente de nuevo.")

print("\nPrograma terminado")
print("Creador de software por Samuel Martinez NL=35")
```
