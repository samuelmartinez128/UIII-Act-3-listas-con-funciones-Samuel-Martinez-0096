#### LISTAS CIENTIFICO

```python
# Listas para almacenar los datos (simulando 3 espacios)
nombres = ["Vacío", "Vacío", "Vacío"]
claves = ["0", "0", "0"]
especialidades = ["N/A", "N/A", "N/A"]

# --- Definición de Funciones ---

def agregarcientifico(indice):
    print(f"\n--- Registrando Científico en posición {indice + 1} ---")
    nombres[indice] = input("Nombre: ")
    claves[indice] = input("Clave: ")
    especialidades[indice] = input("Especialidad: ")
    print(f"[OK]: {nombres[indice]} agregado correctamente.")

def leercientifico(indice):
    print(f"\n--- Datos en posición {indice + 1} ---")
    print(f"Nombre: {nombres[indice]}")
    print(f"Clave: {claves[indice]}")
    print(f"Especialidad: {especialidades[indice]}")

def buscarcientifico():
    busqueda = input("\nIngrese la clave a buscar: ")
    if busqueda in claves:
        pos = claves.index(busqueda)
        print(f"[ENCONTRADO]: La clave pertenece a {nombres[pos]} ({especialidades[pos]})")
    else:
        print("[ERROR]: Clave no encontrada.")

def modificarcientifico(indice):
    print(f"\nModificando a: {nombres[indice]}")
    nombres[indice] = input("Nuevo nombre: ")
    especialidades[indice] = input("Nueva especialidad: ")
    print("[SISTEMA]: Datos actualizados.")

def borrarcientifico(indice):
    print(f"\n[AVISO]: Borrando datos en posición {indice + 1}...")
    nombres[indice] = "Vacío"
    claves[indice] = "0"
    especialidades[indice] = "N/A"
    print("[SISTEMA]: Registro limpiado.")

# --- Menú Principal con While True ---

while True:
    print("\n" + "="*30)
    print("   SISTEMA DE LISTAS CIENTÍFICAS")
    print("="*30)
    print("1. Agregar Científico 1")
    print("2. Leer Científico 1")
    print("3. Buscar por Clave")
    print("4. Modificar Científico 3")
    print("5. Borrar Científico 1")
    print("6. SALIR")

    opcion = input("\nSeleccione una opción: ")

    if opcion == "1":
        agregarcientifico(0) # El índice 0 es el primer lugar

    elif opcion == "2":
        leercientifico(0)

    elif opcion == "3":
        buscarcientifico()

    elif opcion == "4":
        modificarcientifico(2) # El índice 2 es el tercer lugar

    elif opcion == "5":
        borrarcientifico(0)

    elif opcion == "6":
        print("\nSaliendo del programa de listas...")
        break

    else:
        print("\n[ERROR]: Opción no válida.")

print("\nPrograma terminado")
print("Creador de software por Samuel Martinez NL=35")
```
